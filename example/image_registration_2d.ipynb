```python
import rtk
```

```python
fixed_img = rtk.image.ScalarImage(filename='data/square.nii')
moving_img = rtk.image.ScalarImage(filename='data/circle.nii')
similarity = rtk.similarity.SSD(penalty=1000.)
regularizer = rtk.regularizer.BiharmonicRegularizer(convexity_penalty=1., norm_penalty=1.)
```

```python
reg = rtk.registration.LDDMM(
        ndim=fixed_img.ndim,
        n_step=32,
        similarity=similarity,
        regularizer=regularizer,
        n_iters=(50, 20, 10),
        resolutions=(4, 2, 1),
        smoothing_sigmas=(2, 1, 0),
        delta_phi_threshold=1.,
        unit_threshold=0.1,
        learning_rate=0.1,
        parallel=False)
```

```python
reg.set_images(fixed_img, moving_img)
```

```python
warp = reg.execute()
```

```python
warped_moving_img = moving_img.apply_transform(warp)
```

```python
moving_img.show()
warped_moving_img.show()
fixed_img.show()
```
